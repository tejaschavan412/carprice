```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("clean_data.csv")
```

```python
df.info()
```

```python
len(df['name'].unique())
```

```python
len(df['company'].unique())
```

```python
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder()
```

```python
ohe.fit(df[['name','company','fuel_type']])
```

```python
ohe.categories_
```

```python
ct =make_column_transformer((OneHotEncoder(categories=ohe.categories_),["name","company","fuel_type"]),remainder="passthrough")
```

```python
ct
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
```

```python
from sklearn.pipeline import make_pipeline
```

```python
pipe = make_pipeline(ct,reg)
```

```python
pipe
```

```python
from sklearn.model_selection import train_test_split
```

```python
df.columns
```

```python
X= df[['name', 'company', 'year',  'kms_driven', 'fuel_type']]
X
```

```python
y = df[['Price']]
y
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.1,random_state=0)
```

```python
X_train
```

```python
pipe.fit(X_train,y_train)
```

```python
input = pd.DataFrame(data=[['Grand i10 Magna 1.2 Kappa VTVT','Hyundai',2015,30000,'Diesel']],columns=['name', 'company', 'year',  'kms_driven', 'fuel_type'])
```

```python
input
```

```python
pipe.predict(input)
```

```python
ypred = pipe.predict(X_test)
ypred = pd.DataFrame(data=ypred,columns=['Result'])
ypred
```

```python
y_test
```

```python
result = pd.concat([y_test.reset_index(drop= True),ypred.reset_index(drop= True)],axis=1)
result
```

```python
from sklearn.metrics import r2_score
score = r2_score(result['Price'],result['Result'])
score
```

```python
arr =[]
for i in range(0,101):
    X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.1,random_state=i)
    pipe.fit(X_train,y_train)
    ypred = pipe.predict(X_test)
    ypred = pd.DataFrame(data=ypred,columns=['Result'])
    result = pd.concat([y_test.reset_index(drop= True),ypred.reset_index(drop= True)],axis=1)
    score = r2_score(result['Price'],result['Result'])
    arr.append(score)
```

```python
arr
```

```python
index = np.argmax(arr)
index
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.1,random_state=90)
pipe.fit(X_train,y_train)
ypred = pipe.predict(X_test)
ypred = pd.DataFrame(data=ypred,columns=['Result'])
result = pd.concat([y_test.reset_index(drop= True),ypred.reset_index(drop= True)],axis=1)
score = r2_score(result['Price'],result['Result'])
score*100
```
