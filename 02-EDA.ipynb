```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("clean_data.csv")
```

```python
df.info()
```

```python
company_Count = df['company'].value_counts()
```

```python
company_Count.plot(kind='bar',figsize=(12, 6), color='skyblue')
plt.title("Number of Vehicles per Company")

plt.xlabel("Company")

plt.ylabel("Vehicle Count")

plt.xticks(rotation=45)
for i, value in enumerate (company_Count):

    plt.text(i, value + 1, str(value), ha='center', va='bottom', fontsize=9)

    plt.tight_layout()


plt.show()
```

```python
yearcount = df['year'].value_counts()
```

```python
yearcount.plot(kind='bar',figsize=(12, 6), color='skyblue')
plt.title("Number of Vehicles per Year")

plt.xlabel("Year")

plt.ylabel("Vehicle Count")

plt.xticks(rotation=45)
for i, value in enumerate (yearcount):

    plt.text(i, value + 1, str(value), ha='center', va='bottom', fontsize=9)

    plt.tight_layout()


plt.show()
```

```python
fuel_type_count = df['fuel_type'].value_counts()
fuel_type_count
```

```python
fuel_type_counts = df["fuel_type"].value_counts()

plt.figure(figsize=(8, 8))

plt.pie(
    fuel_type_counts,
    labels=fuel_type_counts.index,
    autopct="%1.1f%%",
    startangle=90,
    colors=plt.cm.Set3.colors,   # nice color palette
    explode=[0.05]*len(fuel_type_counts),  # slight separation for all slices
)

plt.title("Distribution of Vehicles by Fuel Type", fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))

sns.swarmplot(data=df, x='year', y='Price')

plt.title("Swarm Plot: Fuel Type vs Price")

plt.xlabel("Fuel Type")

plt.ylabel("Vehicle Price")

plt.xticks(rotation=45)

plt.tight_layout()

plt.show()
```
